```python
import requests
import pandas as pd
```

```python
url = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 10,
    "page": 1
}

response = requests.get(url, params=params)

data = response.json()
```

```python
df = pd.DataFrame(data)

df.head()
```

```python
df = df[[
    "name",
    "symbol",
    "current_price",
    "market_cap",
    "total_volume"
]]

df.head()
```

```python
from datetime import datetime

df["timestamp"] = datetime.now()

df.head()
```
